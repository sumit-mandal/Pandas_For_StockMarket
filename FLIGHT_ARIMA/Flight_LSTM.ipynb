```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('dark_background')
import pandas as pd
import math
```

```python
dataframe = pd.read_csv('FlightCSV/AirPassengers.csv',usecols=[1])
dataframe
```

```python
dataframe = dataframe.rename(columns={'#Passengers':'Passengers'})
dataframe
```

```python
dataframe.plot(legend=True)
```

```python
#Convert pandas dataframe to numpy array
dataset = dataframe.to_numpy()
dataset = dataframe.astype('float32') #convert values to float
print(type(dataset))
print(dataset)
```

### Normalising 

```python
#Scaling values between 0-1
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler((0,1))
dataset = scaler.fit_transform(dataset)
print(dataset)
```

 - We cannot use random way of splitting dataset into train and test asthe sequence of events is important for time series.So let us take first 60% values for train and the remaining 1/3 for testing

## Splitting training and test data

```python
train_size = int(len(dataset)*0.66)
test_size = int(len(dataset)-train_size)
train = dataset[0:train_size,:]
test = dataset[train_size :len(dataset),:]
```

```python
print("train data\n",train.transpose())
print("test data\n",test.transpose())
```

- seq_size is the number of previous time steps to use as input variables to predict the next time period.
- create a dataset where X is the number of passengers at a given time (t, t-1, t-2...) . And Y is the number of passengers at the next time (t + 1).

```python
def to_sequences(dataset,seq_size=1):
    x=[]
    y=[]
    
    for i in range(len(dataset)-seq_size-1):
#         print("i value is\n",i)
        window = dataset[i:(i+seq_size),0]
        x.append(window)
        y.append(dataset[i+seq_size, 0])
        
    return np.array(x),np.array(y)
```

```python
seq_size = 20
# our value contains data with previous 20 time steps and y label is 20th value
```

```python
trainX, trainY = to_sequences(train, seq_size)
testX, testY = to_sequences(test, seq_size)   #Here train and test are, train = dataset[0:train_size,:] test = dataset[train_size :len(dataset),:]
```

```python
np.set_printoptions(threshold=np.inf)
print("trainX is \n",trainX)
print("testX is \n",testX)
print("trainY is \n",trainY)
print("testY is \n",testY)
```

```python
print("Shape of training set{}".format(trainX.shape))
print("Shape of test set{}".format(testX.shape))
```

```python
#reshaping oour array
trainX = np.reshape(trainX,(trainX.shape[0],1,trainX.shape[1]))
testX = np.reshape(testX,(testX.shape[0],1,testX.shape[1]))
```

```python
print("trainX shape:",trainX.shape)
print("testX shape:",testX.shape)
```

## Creating LSTM Layer

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM, Flatten
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
#from keras.callbacks import EarlyStopping
from keras.layers import ConvLSTM2D
```

```python
model = Sequential()
model.add(LSTM(50,activation='relu',return_sequences=True,input_shape=(None,seq_size)))
model.add(LSTM(50,activation='relu'))
model.add(Dense(32)) #hidden dense layer
model.add(Dense(1)) #Output layer
model.compile(optimizer='adam',loss='mse')
```

```python
#fitting the model
model.fit(trainX,trainY,validation_data=(testX,testY),verbose=1,epochs=100)
```

```python
trainPredict = model.predict(trainX)
testPredict = model.predict(testX)
```

```python
# invert predictions back to prescaled values
#This is to compare with original input values
#SInce we used minmaxscaler we can now use scaler.inverse_transform
#to invert the transformation.
trainPredict = scaler.inverse_transform(trainPredict)
trainY = scaler.inverse_transform([trainY])
testPredict = scaler.inverse_transform(testPredict)
testY = scaler.inverse_transform([testY])
```

```python
#calculate root meean squared
trainScore = math.sqrt(mean_squared_error(trainY[0],trainPredict[:,0]))
print('Train Score: %2f RMSE' %(trainScore))
```

```python
testScore = math.sqrt(mean_squared_error(testY[0],testPredict[:,0]))
print('Test Score: %2f RMSE' %(testScore))
```

```python
# shift train prediction for plotting
#we must shift the predictions so that they align on the x-axis with the original dataset.
trainPredictPlot = np.empty_like(dataset)
trainPredictPlot[:,:] = np.nan
trainPredictPlot[seq_size:len(trainPredict)+seq_size,:] = trainPredict
```

```python
#Shift test predictions for plotting
testPredictPlot = np.empty_like(dataset)
testPredictPlot[:,:] = np.nan
testPredictPlot[len(trainPredict)+(seq_size*2)+1:len(dataset)-1,:] = testPredict
```

```python
# plot baseline and predictions
plt.plot(scaler.inverse_transform(dataset))
plt.plot(trainPredictPlot)
plt.plot(testPredictPlot)
plt.show()
```
