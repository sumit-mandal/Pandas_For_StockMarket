# Time shifting

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('walmart_stock.csv',index_col='Date',parse_dates=True)
```

```python
df.head()
```

```python
df.tail()
```

```python
df.shift(periods=1) #shift one index upwards

```

```python
df.shift(periods=-1) #shifts backwards
```

```python
df.shift(freq='M').head()

#it shifts or arranges the value of whole month to the end of the month i.e. 31 january
```
